```python
import numpy as np
import notebooks.scripts.geometrical_objects as geos
import plotly.express as px
import pandas as pd
```

```python
spiral = geos.Spiral(0, 2, t_max=6)
```

```python
def normalizer(spiral):
	def _normalizer(t):
		return np.array([spiral.inv_length(spiral.length*tt/spiral.t_max) for tt in t])
	return _normalizer
```

```python
num_samples = 10_000
t = spiral.t_max*np.random.rand(num_samples)
```

```python
t_norm = normalizer(spiral)(t)
```

```python
x1, x2 = spiral(t).T.tolist()
l = spiral.length_at(t)
samples = pd.DataFrame({
    'time': t,
    'x': x1,
    'y': x2,
	'l': l
})

x1, x2 = spiral(t_norm).T.tolist()
l = spiral.length_at(t_norm)
samples_norm = pd.DataFrame({
    'time': t,
    'x': x1,
    'y': x2,
    'l': l
})

samples['Type'] = 'Original'
samples_norm['Type'] = 'Normalized'

df = pd.concat([samples, samples_norm])
```

```python
fig = px.scatter(df, x='x', y='y', color ='Type', width = 1300, height=500, color_discrete_sequence=['#CC7777', '#77CC77', '#7777CC'], opacity = 0.7, facet_col='Type')
fig.update_layout(showlegend = False)
fig.show()
```
